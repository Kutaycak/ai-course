```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x=df.drop('Outcome',axis=1)
y=df[['Outcome']]
```

```python
model=Sequential()
model.add(Dense(8,activation='relu'))
model.add(Dense(120,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=250,batch_size=32,validation_split=0.20,verbose=1 )
```

```python
# DNN - Regression
```

```python
df=pd.read_pickle('kc_house.pkl')
```

```python
df.head()
```

```python
x=df.drop('price',axis=1)
y=df[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(120,activation='relu'))
model.add(Dense(80,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1)) #regression
model.compile(loss='mse',optimizer='adam')
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=300)
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test,tahmin)
```

```python
mean_squared_error(y_test,tahmin)**.5
```
